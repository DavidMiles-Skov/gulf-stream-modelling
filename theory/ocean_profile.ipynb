```python
import numpy as np
import matplotlib.pyplot as plt

# Stommel's one-box model

tau = np.arange(0, 100, 0.1)
x_tau_1 = np.array([0, 3/8])
x_tau_2 = np.array([5/8, 1])

x = lambda t: 1 - np.exp(-t/6)
y = lambda t: 1 - np.exp(-t)

y_sigma = lambda sigma, x_t: 2*x_t -sigma

markers = np.array([0,1,2,3,4,5,100])

# Temperature vs Salinity, green graph
plt.plot(x(tau), y(tau), c='orange')

plt.xlim(0,1)
plt.ylim(0,1)

# Denisity anomaly annotations
plt.annotate('\u03C3=-1/2', xy=(1/8, 3/4), rotation=60)
plt.annotate('\u03C3=0', xy=(1/4, 1/2), rotation=60)
plt.annotate('\u03C3=1/2', xy=(1/2, 1/2), rotation=60)
plt.annotate('\u03C3=1', xy=(3/4, 1/2), rotation=60)
plt.annotate('\u03C3=3/2', xy=(7/8, 1/4), rotation=60)



# Density anomalies, represented by y = 2*x - sigma
# make lines dashed
# Leave empty space between lines and annotations
# sigma = -1/2, 0, 1/2, 1, 3/2

plt.plot(x_tau_1, y_sigma(-1/2, x_tau_1), c='k', linestyle='dashed')
plt.plot(x_tau_2, y_sigma(-1/2, x_tau_2), c='k', linestyle='dashed')



plt.plot(x_tau_1, y_sigma(0, x_tau_1), c='k', linestyle='dashed')

plt.plot(x_tau_1, y_sigma(1/2, x_tau_1), c='k', linestyle='dashed')
plt.plot(x_tau_1, y_sigma(1, x_tau_1), c='k', linestyle='dashed')
plt.plot(x_tau_1, y_sigma(3/2, x_tau_1), c='k', linestyle='dashed')


# tau annotations

plt.annotate('\u03C4  \u2192 \u221E', xy=(0.88, 0.95))
plt.annotate('\u03C4=0', xy=(0.04, 0.025))
plt.annotate('1', xy=(0.168, 0.624))
plt.annotate('2', xy=(0.303, 0.840))
plt.annotate('3', xy=(0.401, 0.912))
plt.annotate('4', xy=(0.498, 0.948))
plt.annotate('5', xy=(0.57, 0.952))



# Time markers, orange color
plt.scatter(x(markers), y(markers), marker='o', c='orange')



# make dpi = 600
plt.savefig('one_box_xy.png', dpi=600)

plt.xlabel('Salinity')
plt.ylabel('Temperature')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Stommel's one-box model
tau = np.arange(0, 100, 0.1)

x = lambda t: 1 - np.exp(-t/6)
y = lambda t: 1 - np.exp(-t)

y_sigma = lambda sigma, t: 2 * x(t) - sigma

markers = np.array([0, 1, 2, 3, 4, 5, 100])

# Temperature vs Salinity, green graph
plt.plot(x(tau), y(tau), c='orange')

plt.xlim(0, 1)
plt.ylim(0, 1)

# Density anomaly annotations
density_labels = [r'$\sigma=-\frac{1}{2}$', r'$\sigma=0$', r'$\sigma=\frac{1}{2}$', r'$\sigma=1$', r'$\sigma=\frac{3}{2}$']
density_values = [-1/2, 0, 1/2, 1, 3/2]
positions = [1/8, 1/4, 1/2, 3/4, 7/8]

for label, sigma, pos in zip(density_labels, density_values, positions):
    plt.annotate(label, xy=(pos, y_sigma(sigma, 0) + 0.05), rotation=60, color='b')

# Density anomalies, represented by y = 2*x - sigma
# Leave empty space between lines and annotations
# Adjust the offset to leave space for annotations
sigma_levels = [-1/2, 0, 1/2, 1, 3/2]
offset = 0.01  # Adjust this value to control the separation between lines and annotations

for sigma in sigma_levels:
    y_sigma_values = y_sigma(sigma, tau)
    plt.plot(x(tau), y_sigma_values - offset, c='b', linestyle='dashed')

# tau annotations
plt.annotate('\u03C4  \u2192 \u221E', xy=(0.88, 0.95))
plt.annotate('\u03C4=0', xy=(0.04, 0.025))
plt.annotate('1', xy=(0.168, 0.624))
plt.annotate('2', xy=(0.303, 0.840))
plt.annotate('3', xy=(0.401, 0.912))
plt.annotate('4', xy=(0.498, 0.948))
plt.annotate('5', xy=(0.57, 0.952))

# Time markers, orange color
plt.scatter(x(markers), y(markers), marker='o', c='orange')

# Make dpi = 600
# plt.savefig('one_box_xy.png', dpi=600)

plt.xlabel('Salinity')
plt.ylabel('Temperature')
plt.show()
```
