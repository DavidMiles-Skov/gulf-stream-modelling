### 3 Box-Model

- Same temperature in two polar boxes
- negligible boundary salinity effects

```python
# Imports

import matplotlib.pyplot as plt
import numpy as np
from phaseportrait import PhasePortrait2D, PhasePortrait3D
```

Testing with symmetrical salinity flux

```python

psi=0.5
gamma=0.5

def dF_three_box(x_N, x_L, psi=psi, gamma=gamma):

    x_N = gamma*psi-abs(1-x_L)*x_L-2*abs(1-x_N)*x_N
    x_L = (1-gamma)*psi-abs(1-x_N)*x_N-2*abs(1-x_L)*x_L
    
    return x_N, x_L

plt.rcParams['figure.figsize'] = [10, 10]
plt.rcParams['figure.dpi'] = 600

three_box_reduced_portrait = PhasePortrait2D(dF_three_box, [-5,5], MeshDim=30, Title=r'', xlabel=r"$x_N$", ylabel=r"$x_L$")
fig, ax = three_box_reduced_portrait.plot()
plt.show()
# save figure
# fig.savefig(f"gif1/kd_{kd}.png")
```
