```python
from phaseportrait import PhasePortrait2D, PhasePortrait3D
import numpy as np # We are also going to use it, for an example involving trigonometric functions
```

```python
# Increasing the size of the figures
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
# increasing dpi
plt.rcParams['figure.dpi'] = 600
```

```python
ω = 1.00

def dFOscillator(x, y):
    return y, -ω*ω*x

def dFOscillator(x, y, *, ω=1):
    return y, -ω*ω*x

Oscillator = PhasePortrait2D(dFOscillator, [-5,5])
Oscillator.plot()
```

```python
Oscillator2 = PhasePortrait2D(dFOscillator, [-5,5], MeshDim=21)
Oscillator2.plot()
```

```python
def dFPolar(r, θ):
    return 0.5*r*(1 - r*r), 1+0*θ


PolarCoordinates = PhasePortrait2D(dFPolar, [-3, 3], Polar=True, Title='Limit cycle')
PolarCoordinates.plot()
```

```python
def dFDampedPendulum(theta, v):
    return v, -0.75*v-np.sin(theta)

DampedPendulum = PhasePortrait2D(dFDampedPendulum, [-4, 4], MeshDim=100, Title='Damped pendulum', xlabel=r"$\Theta$", ylabel=r"$\dot{\Theta}$")
DampedPendulum.plot()
```

```python
import matplotlib.pyplot as plt

import phaseportrait
from phaseportrait.streamlines import *

def dF(x,y,z):
    return -x, -y, -z

p1 = phaseportrait.PhasePortrait3D(dF, [-3, 3])
p1.plot()
plt.show()



def dF(x,y,z, *, w=1):
    return -y, x, -z + w


p2 = phaseportrait.PhasePortrait3D(dF, [-3, 3], MeshDim=8, maxLen=2500, deltat=0.05)

p2.add_slider('w')

p2.plot(color='viridis', grid=False)
plt.show()
```

### Testing with two-box model

Effect of flow rate on equilibrium salinity and temperature for one box with flow

```python
def one_box(x, y, f=0.01, delta=1/6):
    return delta*(1-x)-abs(f)*x, 1-y-abs(f)*y


one_box_portrait = PhasePortrait2D(one_box, [0, 1], MeshDim=30, Title='One Box-Model (with flow)', xlabel="Salinity", ylabel="Temperature")
one_box_portrait.plot()
```

Two Box-Model

```python
def two_box(x, y):
    # Default values
    R = 2
    delta = 1/6
    lam = 1/5

    return delta*(1-x)-(x/lam)*abs(-y+R*x), 1-y-(y/lam)*abs(-y+R*x)

two_box_portrait = PhasePortrait2D(two_box, [0, 1], MeshDim=30, Title='Two Box-Model', xlabel="Salinity", ylabel="Temperature")
two_box_portrait.plot()

```

Three Box-Model

```python
from phaseportrait import PhasePortrait3D
import matplotlib.pyplot as plt
```

```python
# increasing figure size (width)

plt.rcParams['figure.figsize'] = [20, 10]
plt.rcParams['figure.dpi'] = 600
```

```python
def three_box(x_N, x_L, x_E):
    # Nondimensionalised salinity flux
    psi = -8
    # Nondimensionalised flow rates
    F_L = 1
    F_N = 1

    return (-1/2)*psi + (-x_N-1/2) + abs(F_N)*(x_E-x_N), (-1/2)*psi + (-x_L-1/2) + abs(F_L)*(x_E-x_L), psi + (1-x_E) - abs(F_N)*(x_E-x_N) - abs(F_L)*(x_E-x_L)


three_box_portrait = PhasePortrait3D(three_box, [-5, 5], MeshDim=8,Title=r'', xlabel=r"$x_N$", ylabel=r"$x_L$", zlabel=r"$x_E$")
three_box_portrait.plot()

```
