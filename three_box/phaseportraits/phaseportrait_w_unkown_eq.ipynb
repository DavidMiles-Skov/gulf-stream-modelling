```python
import matplotlib.pyplot as plt
import numpy as np
from phaseportrait import PhasePortrait2D
```

```python
def three_box_dF(x, y, l=-3/4):
    return l-y*abs(1-y)-2*x*abs(1-x), l-x*abs(1-x)-2*y*abs(1-y)
```

```python
# x3, y1
l=-3/4
x3 = 1/2 + np.sqrt(9 - 12*l)/6
y1 = 1/2 + np.sqrt(9 + 12*l)/6
y2 = 1/2 + np.sqrt(9 - 12*l)/6
x1 = 1/2 - np.sqrt(9 + 12*l)/6
x2 = 1/2 - np.sqrt(9 - 12*l)/6
```

Making phaseportait with x3, y1 marked

```python
pp = PhasePortrait2D(three_box_dF,[-2,2], MeshDim=30, Title=r"$\lambda$ = -3/4", xlabel="x", ylabel="y")
pp.plot()
fig, ax = pp.fig, pp.ax
# Adding points
ax.plot(x3, y1, 'o', color='red')
ax.plot(x1, y2, 'o', color='red')
ax.plot(x2, y1, 'o', color='red')

plt.rcParams['figure.figsize'] = [10, 10]
plt.rcParams['figure.dpi'] = 600

plt.show()
```
